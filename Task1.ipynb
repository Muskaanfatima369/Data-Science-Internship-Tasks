```python
#first me import libraries and load data
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Iris Dataset EDA

#load dataset using pandas
df=pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
print("Shape:", df.shape)   #display shape of dataset
print("Columns:", df.columns) #tell how many columns in dataset
print(df.head()) #display first 5 rows
```

```python
#Visualizations of dataset
#Scatter plot to analyze relationships between variables.
plt.figure(figsize=(8, 5))

sns.scatterplot(
    data=df,
    x='sepal_length',
    y='petal_length',
    hue='species',       # Color by species
    palette='Set1'
)

plt.title('Sepal Length vs Petal Length') # puts a heading at the top of graph
plt.xlabel('Sepal Length (cm)') # give name to x-axis
plt.ylabel('Petal Length (cm)') # give name to y-axis
plt.legend(title='Species')
plt.tight_layout()
plt.show()
```

```python
#Histogram to examine data distributio
plt.figure(figsize=(10, 6))

df['sepal_length'].hist(
    bins=20,
    color='steelblue',
    edgecolor='black'
)

plt.title('Distribution of Sepal Length') # puts a heading at the top of graph
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.tight_layout() # prevents labels from being clipped at the edges
plt.show()
```

```python
#Box plot to detect outliers and spread of values.
plt.figure(figsize=(10, 6))

sns.boxplot(
    data=df,
    x='species',
    y='petal_length',
    palette='pastel'
)

plt.title('Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.tight_layout()
plt.show()
```
